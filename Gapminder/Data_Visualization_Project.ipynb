#### Finding Correlation for Life Expectancy

```python
plt.figure(figsize=(12,4))
heatmap_life_exp = sns.heatmap(df_life_exp.corr(), 
                               vmax=1, 
                               vmin=-1, 
                               annot=True, 
                               cmap='BrBG');
heatmap_life_exp.set_title('Life Expectancy Correlation', 
                           fontdict={'fontsize': 11, 'fontweight': 'bold'}, 
                           pad=12);
plt.savefig('heatmap_life_exp.png', dpi=150)
```

```python
plt.figure(figsize=(5,4))
ax = sns.distplot(df_life_exp[2001], kde=True, bins=10, color='Green');
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
plt.xlabel('Life Expectancy in 2001');
plt.savefig("le_2001.png", dpi=150)
```

```python
plt.figure(figsize=(12,5))
ax = sns.pairplot(df_life_exp, dropna=True, hue= 1983)
ax.fig.suptitle("Life Expectancy Over Years", x=0.47, y=1.03, fontsize= 21, fontweight='bold')
plt.savefig('le_pairplot.png', dpi=95)
plt.show()
```

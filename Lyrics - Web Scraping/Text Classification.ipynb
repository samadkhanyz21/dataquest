```python
import pickle
import requests
import time
from bs4 import BeautifulSoup as bs
```

```python
header = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2919.83 Safari/537.36'}
url = 'https://www.azlyrics.com/m/maroon5.html'
r = requests.get(url)
r.status_code
```

```python
soup = bs(r.content, 'lxml')
songs = soup.find_all(class_="listalbum-item")

# Extracting links
urls = []

for s in songs:
    urls.append(s.find('a'))

# Extracting hyperlink
hyperlinks = []

for i in range(len(urls)):
    urls[i]['href'] = 'https://www.azlyrics.com/' + urls[i]['href']
    hyperlinks.append(urls[i]['href'])
    
```

```python
soup_1 = bs(requests.get(hyperlinks[0]).text, "lxml")
soup_1
```

```python
lyrics = []

for link in hyperlinks:
    time.sleep(0.1)
    soup_1 = bs(requests.get(link).text, "lxml")
    div = soup_1.find_all(name='div', attrs={"class":"col-xs-12 col-lg-8 text-center"})
    print(div)
    song_lines = ""
    
    for d in div:
        if d.get("class") != None and "verse" in d.get("class"):
            song_lines = song_lines + d.text
    lyrics.append(song_lines)
```
